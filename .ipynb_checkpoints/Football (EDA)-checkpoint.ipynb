#### Exploratory Data Analysis

```python
v = ['Goals','Assists','Passes_Attempted','Penalty_Goals','Penalty_Attempted','Yellow_Cards','Red_Cards']
for colname in v:
    
    x= df.groupby(['Club'])[colname].sum().sort_values(ascending=True).reset_index(drop=False)
    display(x.style.background_gradient( cmap='flare'))
    sns.barplot(x='Club',y=colname,data=x,palette="flare")
    plt.xticks(ticks=np.arange(0,20,1),labels=x['Club'],fontsize=12,style='oblique',rotation=90)
    plt.show()
```

#### Ratio minutos/partido, goles/minutos 

```python
df['MinsPerMatch'] = (df['Mins']/df['Matches']).astype(int)
df['GoalsPerMatch'] = (df['Goals']/df['Matches']).astype(float)
df.head()
```

#### Goles totales

```python
df['Goals'].sum()
```

#### Penaltis lanzados / Goles de penalti

```python
Goles_penalti = df['Penalty_Goals'].sum()
Penaltis_lanzados = df['Penalty_Attempted'].sum()
```

#### Pie chart penaltis marcados - penaltis fallados

```python
plt.figure(figsize=(10,4))
penalti_no_marcado = Penaltis_lanzados - Goles_penalti
data = [penalti_no_marcado, Goles_penalti]
labels = ['Penaltis fallados', 'Penaltis marcados']
color = sns.color_palette('Set3')
plt.pie(data, labels = labels, colors = color, autopct = '%.0f%%')
plt.show()
```

#### Jugadores por rango de edad

```python
Menores_20 = df[df['Age'] <= 20]
Entre_20_25 = df[(df['Age'] > 20) & (df['Age']<= 25)]
Entre_25_30 = df[df['Age'] > 25 & (df['Age']<= 30)]
Mayores_30 = df[df['Age'] > 30]
```

```python
x = np.array([Menores_20['Name'].count(), Entre_20_25['Name'].count(), Entre_25_30['Name'].count(), Mayores_30['Name'].count()])
mylabels= ('< 20', '> 20 & <= 25', '> 25 & <= 30', '> 30')
color = sns.color_palette('Set3')
plt.pie(x, labels= mylabels, colors=color, autopct= '%.1f%%')
plt.show()
```
